### Import necessary packages

```python
import os
from dotenv import load_dotenv

from sqlalchemy import create_engine

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load environmental variables

```python
os.chdir('../../')
load_dotenv('./src/scraping/config/db_secrets.env')
```

### Create DB connection

```python
hostname = os.getenv("HOSTNAME")
database = os.getenv("DATABASE")
username = os.getenv("USERNAME")
password = os.getenv("PASSWORD")
port = os.getenv("PORT")

db = create_engine(
        f"postgresql://{username}:{password}@{hostname}:{port}/{database}"
    )

team_data = pd.read_sql('select team_name, round, result, goals_for, goals_against, opponent, xg_for, xg_against  from summary_team', con=db)
```

### Team data cleaning

```python
team_data['xg_different'] = team_data['xg_for'] - team_data['xg_against']
team_names = team_data['team_name'].unique()
```

```python
for team in team_names:
    plot_data = team_data[team_data['team_name'] == team]
    fig,ax = plt.subplots(1,1, figsize=(10,5))
    ax.plot(plot_data['round'], plot_data['xg_different'], color = 'navy', label = 'xG Differential')
    ax.set_xticks(list(range(plot_data['round'].min(), plot_data['round'].max()+1)))
    ax.set_yticks(list(range(-5,6)))
    ax.set(title = f"{team}: xG Differential", xlabel = 'Gameweek', ylabel='xG Differential')
    ax.grid(True)
    ax.legend(loc=(1.01,0.45))
```

```python
# Plot xg differential per match week as a scatter plot
for team in team_names:
    plot_data = team_data[team_data['team_name'] == team].reset_index()
    fig,ax = plt.subplots(1,1, figsize=(10,5))
    sns.scatterplot(x = plot_data['round'], y = plot_data['xg_different'], hue=plot_data['result'], palette='Dark2')
    ax.axhline(np.mean(plot_data['xg_different']), linestyle='--', label='Avg xG Differential')
    ax.set_xticks(list(range(plot_data['round'].min(), plot_data['round'].max()+1)))
    ax.set_yticks(list(range(-5,6)))
    ax.set(xlabel = 'Gameweek', ylabel='xG Differential')
    ax.set(title = f"{team}: xG Differential", xlabel = 'Gameweek', ylabel='xG Differential')
    ax.grid(True)
    ax.legend(loc='lower center', bbox_to_anchor=(0.5,-0.25), ncol=4)

    plt.xlim(0, plot_data['round'].max()+1)

    # Add opponent labels
    for i in range(plot_data.shape[0]):
        plt.text(x=plot_data['round'][i]+0.05,y=plot_data['xg_different'][i]+0.01,s=plot_data['opponent'][i], 
                fontsize='small', rotation='vertical')
```
